```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Raw_Data_heart_attack_prediction_dataset")
```

```python
df.drop("Patient ID", axis = 1, inplace = True)
```

```python
df.columns
```

```python
# Mapping dictionary
diet_mapping = {
    'Average': 0,
    'Healthy': 1,
    'Unhealthy': 2
}

# Apply mapping to the column
df['Diet_encoded'] = df['Diet'].map(diet_mapping)
df.drop('Diet', axis = 1, inplace = True)
```

```python
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
df.drop('Blood Pressure', axis = 1, inplace = True)
```

```python
df.drop('Country', axis = 1, inplace = True)
```

```python
df = pd.get_dummies(df, columns=['Continent'])
```

```python
df['hemisphere_encoded'] = df['Hemisphere'].map({'Northern Hemisphere': 1, 'Southern Hemisphere': 0})
df.drop("Hemisphere", axis = 1, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

numerical_features = ['Age', 'Cholesterol', 'Heart Rate', 'BMI', 'Exercise Hours Per Week', 'Stress Level', 'Sedentary Hours Per Day', 
                      'Income', 'Triglycerides', 'Systolic BP', 'Diastolic BP', 
                     'Physical Activity Days Per Week', 'Sleep Hours Per Day', 'Diet_encoded' ]
```

```python
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```
