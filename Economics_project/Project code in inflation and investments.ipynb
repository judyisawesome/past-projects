```python
import pandas_datareader
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt           # for plotting
import datetime as dt     
import os



# We are importing the data method from the package and calling it web. 
import pandas_datareader.data as web    

myfigsize=(15,8)
```

```python
codes = ['CPILFESL']  
                                      

urban = web.DataReader(codes, 'fred', start='1980-01-01')

urban['yearly']=urban["CPILFESL"].pct_change(periods=12)*100
urban.head(20)
urban
```

```python
dj = web.DataReader('^DJI', 'stooq', start='1980-01-01') 
sp = web.DataReader('^SPX', 'stooq',start='1980-01-01') 
ndq= web.DataReader('^NDQ', 'stooq',start='1980-01-01')
```

```python
dj2=dj.resample('m').mean()
dj2['percent']=dj2['Close'].pct_change(periods=12)*100
dj2
```

```python
sp2=sp.resample('m').mean()
sp2['percent']=sp2['Close'].pct_change(periods=12)*100
sp2
```

```python
ndq2=ndq.resample('m').mean()
ndq2['percent']=ndq2['Close'].pct_change(periods=12)*100
ndq2
```

```python


fig, ax1 = plt.subplots(figsize=(20,10))

color = 'tab:red'
ax1.set_xlabel('Date',size = 20)
ax1.set_ylabel('Stock price change(percentage)',size = 20)
ax1.set_title('Stock Performance Since 1980',size = 25)

ax1.plot(ndq2.index, ndq2['percent'], color='grey',label="Nasdaq", alpha=0.7)
ax1.plot(sp2.index, sp2['percent'], color='blue',label="S&P 500")
ax1.plot(dj2.index, dj2['percent'], color='pink',label="Dow Jones Industrail average", alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.legend()
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)


ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('inflation(percent)', color=color,size = 20)
ax2.plot(urban.index, urban['yearly'], color='red',label='inflation')
ax2.legend()
ax2.tick_params(axis='y', labelcolor=color)
ax2.axvspan(dt.datetime(2021,1,1), dt.datetime(2022,5,1),color="green", alpha=0.1)

fig.tight_layout()
plt.savefig('stockindex.svg')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.rcParams['font.size'] = '15'
plt.show()


```

```python
treasury = web.DataReader(['DGS10'], 'fred', start='1980-01-01')
treasury2=treasury.resample('m').mean()
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
color = 'tab:blue'
ax1.plot(treasury2.index, treasury2['DGS10'], color='blue', label='Treasury')


ax1.set_ylabel('bond yield',size=20)
ax1.set_xlabel('Date',size=20)
ax1.set_title('US Bond Yields Since 1980',size=25)

ax1.tick_params(axis='y', labelcolor=color)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.legend()
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.text(dt.datetime(1985,3,1), 12, "US Bond Yields",size=18)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)


ax2 = ax1.twinx()


color = 'tab:red'
ax2.set_ylabel('inflation(percent)', color=color,size=20)
ax2.plot(urban.index, urban['yearly'], color='red',label='inflation')
ax2.legend()
ax2.tick_params(axis='y', labelcolor=color)
ax2.text(dt.datetime(1985,4,1), 2, "Inflation",size=20
        )
ax2.axvspan(dt.datetime(2021,1,1), dt.datetime(2022,5,1),color="green", alpha=0.1)

fig.tight_layout()
plt.savefig('bond.svg')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.rcParams['font.size'] = '16'
plt.show()
```

```python
gold=pd.read_excel('Prices.xlsx',header=8,sheet_name='Daily',index_col=3)
gold=gold[['US dollar']]
gold.index.rename('Date',inplace = True)
gold2=gold.resample('m').mean()

gold2
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))

color = 'tab:blue'
ax1.set_xlabel('Date',size=20)
ax1.set_ylabel('gold price(dollars)',size=20)
ax1.set_title('Gold Price Since 1980', size = 25)
ax1.plot(gold2.index, gold2['US dollar'], color='blue',label='Gold Price')
ax1.tick_params(axis='y', labelcolor=color)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.text(dt.datetime(2010,12,1), 1800, "Gold Price",size=20)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('inflation(percent)', color=color,size=20)
ax2.plot(urban.index, urban['yearly'], color='red',label='inflation')

ax2.tick_params(axis='y', labelcolor=color)
ax2.text(dt.datetime(2015,1,1), 1, "Inflation",size=20)
ax2.axvspan(dt.datetime(2021,1,1), dt.datetime(2022,5,1),color="green", alpha=0.1)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

plt.savefig('gold.svg')
plt.show()
```
